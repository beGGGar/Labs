```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def f2(x):
    return x**3/3+1

def eiler(f, x0,y0,b,n):
    x = symbols('x')
    y = symbols('y')
    h=(b-x0)/n
    xi=[x0]
    yi=[y0]
    fi=[]
    for i in range(n):
        fi=fi+[(f.subs(y,yi[i])).subs(x,xi[i])]
        xi = xi+[xi[i]+h]
        yi = yi+[yi[i]+h*fi[i]]
    u=yi[n]
    x1 = np.arange(0, 3, 0.02)

    plt.plot(xi, yi,x1,f2(x1),'r')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$f(x)$')
    plt.grid(True)
    plt.show()
    return u
    
x = symbols('x')    
f=x**2    
x0=0;
y0=1;
X=3;
n=50;
u=eiler(f,x0,y0,X,n);

dy=np.abs(u-(3**3/3+1))
print(dy)

```

В качестве отрезка для численного решения ОДУ выбрали [0 3], отрезок было принято
разделить на 50 частей. На графике синяя кривая численное решение ОДУ, тогда как
красная точное решение. Абсолютная погрешность составила 0.2682
